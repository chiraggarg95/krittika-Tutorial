```python
file_name='Moons_and_planets.csv'

file=open(file_name, 'r')
lines=file.read().split('\n')

del(lines[0])                                     #removing title line
del(lines[-1])                                    #removing last empty line
```

```python
moon_to_planet_dict=dict()       #to give the planet of required moon
moon_to_dist_dict=dict()        #to give dist of required moon
moon_count=dict()                  #to give moons of required planet
```

```python
#collecting data

for line in lines:
    moon, planet, dist = line.split(',')[0], line.split(',')[1], line.split(',')[2]
    
    moon_to_planet_dict[moon]=planet
    moon_to_dist_dict[moon]=float(dist)
    moon_count[planet]=moon_count.get(planet, 0)+1
```

```python
print(moon_count)             #moons of each planet.
```

```python
#sorting moons by dist

sorted_moon_vs_dist = sorted([(dist, moon) for (moon, dist) in moon_to_dist_dict.items() ], reverse=True)
print('Ordered List is: \n')
for i, entry in enumerate(sorted_moon_vs_dist):
    print('Rank: {}      Moon: {}        Planet: {}       Distance: {}'.format(i+1, entry[1], moon_to_planet_dict[entry[1]], entry[0]))
```
